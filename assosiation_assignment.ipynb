```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
#!pip install mlxtend
import matplotlib.pyplot as plt
import numpy as np
```

```python
book=pd.read_csv("/content/book.csv")
```

```python
movies=pd.read_csv("/content/my_movies.csv")
```

```python
book
```

```python
#apriori algorithm
```

```python
frequent_itemsets = apriori(book, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
#for different support values
```

```python
frequent_itemsets1 = apriori(book, min_support=0.2, use_colnames=True)
frequent_itemsets1
```

```python
rules1 = association_rules(frequent_itemsets1, metric="lift", min_threshold=0.7)
rules1
```

```python
frequent_itemsets2 = apriori(book, min_support=0.25, use_colnames=True)
frequent_itemsets2
```

```python
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=0.7)
rules2
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
