```python
plt.figure(figsize=(12,8))
heatmap_y_month = pd.pivot_table(data=rate,values="inflation Rate (%)",index="Annual Change",columns="Year",aggfunc="mean",fill_value=0)
sns.heatmap(heatmap_y_month,annot=True,fmt="g") #fmt is format of the grid values
```

```python
rate.columns
```

```python
# Checking Data Distribution
sns.histplot(rate['inflation Rate (%)'])
plt.show()
```

```python
# Checking Outlir
plt.boxplot(x='inflation Rate (%)',data=rate)
plt.show()
```

```python
# Remoove the outlier 
rate.drop(rate.index[rate['inflation Rate (%)']<-1],inplace=True)
rate.drop(rate.index[rate['inflation Rate (%)']>17],inplace=True)
```

```python
plt.boxplot(x='inflation Rate (%)',data=rate)
plt.show()
```

```python
plt.figure(figsize=(12,3))
sns.lineplot(x="Annual Change",y="inflation Rate (%)",data=rate)
```
